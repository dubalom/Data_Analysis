```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
def perform_clustering(file_path, n_clusters=3):
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # Select features for clustering (excluding 'Year')
    features = ['GDP', 'Inflation', 'Unemployment', 'Exports']
    X = df[features]
    
    # Apply KMeans clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df['Cluster'] = kmeans.fit_predict(X)
    
    # Plot clusters using GDP vs. Exports
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='GDP', y='Exports', hue='Cluster', palette='viridis', s=100)
    plt.title("Clustering of Economic Data")
    plt.xlabel("GDP (Billion USD)")
    plt.ylabel("Exports (Billion USD)")
    plt.savefig("../plots/economic_clusters.png")
    plt.show()
    
    print("Clustering complete. Cluster labels have been added to the dataset.")

if __name__ == "__main__":
    file_path = "../data/russian_economic_data.csv"
    perform_clustering(file_path)
```
